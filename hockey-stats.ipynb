```python
#thanks to https://pypi.org/project/nhl-api-py/ for the API source!
#pip install nhl-api-py
import pandas as pd
print(pd.__version__)

```

```python
from nhlpy import NHLClient
```

```python
client = NHLClient(verbose=True)
teams = client.teams.teams_info()
```

```python
#test of output
print(teams)
```

```python

def clean_name(ntype, name):
    """
    Clean the name of the player or team
    """
    return name[ntype]['default']

df = pd.DataFrame()

for team in teams:
    players = client.teams.roster(team_abbr=team['abbr'], season="20242025")
    for p in players['forwards'] + players['defensemen'] + players['goalies']:
        p['team'] = team['abbr']
        p['firstName'] = clean_name('firstName', p)
        p['lastName'] = clean_name('lastName', p)
        p['player_name'] =  p['firstName'] + ' ' +  p['lastName']
    forwards, defense, goalies = players['forwards'], players['defensemen'], players['goalies']

    df = pd.concat(
        [df, 
         pd.DataFrame(forwards), 
         pd.DataFrame(defense), 
         pd.DataFrame(goalies)
         ], ignore_index=True)
```

```python
# Create Pygal Treemaps for both 2024 and 1992 Olympics
import pygal
from pygal.style import Style
# Define custom styles for the treemaps
custom_style_2024 = Style(
    #colors=['#009BFF', '#22A8FF', '#35B0FF', '#4FBAFF', '#62C1FF', '#78CAFF', '#8FD3FF','#A5DCFF', '#B9E4FF', '#CCEBFF', '#DEF1FC','#E52300'],
    value_font_size=14,
    label_font_size=16,
)
treemap_2024 = pygal.Treemap(style=custom_style_2024,legend_at_bottom=True)
treemap_2024.title = 'NHL Teams by Country for 2024-2025 Season'
```

```python
treemap_2024.render_to_file('2024_2025_nhl_teams_treemap.svg')
```

```python
# Count the number of players per team per country
team_country_counts = df.groupby(['birthCountry', 'team', 'lastName']).size().reset_index(name='Count')
# Populate the treemaps with data, grouped by team within each country
for country in team_country_counts['birthCountry'].unique():
    country_data = team_country_counts[team_country_counts['birthCountry'] == country]
    country_node = []
    for team in country_data['team'].unique():
        team_data = country_data[country_data['team'] == team]
        combined_label = f"{team} - {', '.join(team_data['lastName'])}"
        team_count = team_data['Count'].sum()  # Summing the counts for each team
        country_node.append({'label': combined_label, 'value': team_count})
    treemap_2024.add(country, country_node)
```

```python
treemap_2024.render_to_file('2024_2025_nhl_teams_treemap.svg')
```
